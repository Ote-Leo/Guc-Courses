### What is the medal distribution of the regions?

```python
Variables['Medals_Region_Distribution'] = Data_Merged.groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending = False)
Variables['Medals_Region_Distribution']
```

By checking [external sources](https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table) a difference appears in the total number of medal won by countries & the previous output.

The difference in values might be that medal counts are per each athlete rather than the teams.

```python
Variables['Unique_Medals'] = Data_Merged.drop_duplicates(subset=['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
Variables['Medals_Region_Distribution'] = Variables['Unique_Medals'].groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False)
Variables['Medals_Region_Distribution']['Total'] = Variables['Medals_Region_Distribution']['Gold'] + Variables['Medals_Region_Distribution']['Silver'] + Variables['Medals_Region_Distribution']['Bronze']

# Viewing a tabular form of the top 15 countries
Variables['Medals_Region_Distribution'].iloc[:15].style.set_caption('Medals by Country Over the past 120 years [Top 15]')\
    .bar(subset=['Gold'], color='#f0c05a')\
    .bar(subset=['Silver'], color='Lightgray')\
    .bar(subset=['Bronze'], color='#a97142')
```

Now the output does match the values in the [external sources](https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table) with the exception of minute differences in the total medals.

The difference might be a problem in the dataset.

Plotting the top 10 Countries with medal distribution

```python
Variables['Medals_Region_Distribution'][['Gold', 'Silver', 'Bronze']].head(15).plot(kind = 'barh', stacked = True, figsize = (10, 10))

plt.xlabel('Medals Count', fontsize = 12, weight = 'bold')
plt.ylabel('Regions', fontsize = 12, weight = 'bold')
plt.title('Top 15 Countries - Medal Distribution', fontsize = 20, weight = 'bold')
plt.legend(bbox_to_anchor = (1, 1), loc = 2)
sns.despine()

plt.show()
```

## Median Height/Weight of the Olympic medalists

Let's try to plot a scatterplot of height vs weight to see the distriubtion of values (without group by sport). We'll be using the Data_GoldMedals_Ages dataframe

```python
Variables['Data_GoldMedals_Ages'].info()
```

We can see that we have some missing values in the height and weight columns.

At this point, we can:

1. Use only the rows that has a value in the Height & Weight columns
2. Replace the values with the mean of the column

The second solution might lead to misleading results: we are talking about data of athletes of different ages and different sports (that have done different training), and a solution for this will be too complicated.

Thus, we'll be using the $1^{st}$ appraoch

First we'll create a dataframe filtering only the rows that has the column Height & Weight populated.

```python
Variables['Data_GoldMedals_Height_Weight'] = Variables['Data_GoldMedals_Ages'][(
    Variables['Data_GoldMedals_Ages']['Height'].notnull() &
    Variables['Data_GoldMedals_Ages']['Weight'].notnull()
)]
```

Quick nullity check

```python
Variables['Data_GoldMedals_Height_Weight'].info()
```

Now that we have 10,000 rows, let's create the scatterplot:

```python
plt.figure(figsize=(12, 10))
ax = sns.scatterplot(x='Height', y='Weight', data=Variables['Data_GoldMedals_Height_Weight'])
plt.title('Height vs Weight of Olympic Medalists', fontsize = 20, weight = 'bold')
```

The vast majority of teh smaples show a linear relation between height & wight ( the more the weight, the more the height).

We have exceptions & I am willing to know more!

# Constructing a new Data-Set

## New Data-Set Format

| Year     | NOC      | Regoin     | Gold     | Silver     | Bronze     | Total Medals       |
|----------|----------|------------|----------|------------|------------|--------------------|
| $Year_1$ | $NOC_1$  | $Regoin_1$ | $Gold_1$ | $Silver_1$ | $Bronze_1$ | $Total \ Medals_1$ |
| $\vdots$ | $\vdots$ | $\vdots$   | $\vdots$ | $\vdots$   | $\vdots$   | $\vdots$           |
| $Year_n$ | $NOC_n$  | $Regoin_n$ | $Gold_n$ | $Silver_n$ | $Bronze_n$ | $Total \ Medals_n$ |

```python
New_Record = pd.merge(DF_Medals_2021.rename(columns={"Team/NOC" : "region"}), DF_Noc_Regions, on='region').drop(columns=['notes', 'Rank', 'Rank by Total'])
New_Record.loc[:, 'Year'] = 2021
New_Record = New_Record[['Year', 'NOC', 'region', 'Gold', 'Silver', 'Bronze', 'Total']]
New_Record
```

### Constructing the new dataframe

```python
DF_Result          = Data_Merged.groupby(['Year', 'NOC', 'region']).sum().sort_values('Year').reset_index()
DF_Result          = DF_Result.drop(columns = ['ID', 'Age', 'Height', 'Weight'])
DF_Result['Total'] = DF_Result['Gold'] + DF_Result['Silver'] + DF_Result['Bronze']

DF_Result
```

### Adding the new Data frame

```python
DF_Result = DF_Result.append(New_Record, ignore_index=True)
DF_Result
```

```python
DF_Result.groupby('Year').count()['NOC']
```

### Saving the result data frame

```python
DF_Result.to_csv('result_data.csv')
```
