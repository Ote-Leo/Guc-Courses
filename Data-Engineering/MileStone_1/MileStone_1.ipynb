# Overall Analysis & Calculations

```python
Data_Years.insert(0, 'Overall')
Data_Country.insert(0, 'Overall')
```

```python
Medal_DF = Data_Merged.drop_duplicates(subset=['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
Medal_DF
```

Function for getching the data_medal of any country in any year.

```python
def Fetch_Data_Medal(Year, Country):
    X = 0
    
    if Year == 'Overall' and Country == 'Overall':
        Temp_DF = Medal_DF
    
    elif Year == 'Overall' and Country != 'Overall':
        X = 1
        Temp_DF = Medal_DF[Medal_DF['region'] == Country]
    
    elif Year != 'Overall' and Country == 'Overall':
        Temp_DF = Medal_DF[Medal_DF['Year'] == int(Year)]
    else:
        Temp_DF = Medal_DF[(Medal_DF['Year'] == Year) & (Medal_DF['region'] == Country)]
    
    if X == 1:
        Y = Temp_DF.groupby('Year').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Year').reset_index()
    else:
        Y = Temp_DF.groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()
    
    Y['Total'] = Y['Gold'] + Y['Silver'] + Y['Bronze']

    print(Y)
    
```

```python
print('Medal won by Germany in 1900')
Fetch_Data_Medal(1900, 'Germany')

print('\n')

print('Overall Analysis of Countries')
Fetch_Data_Medal('Overall', 'Overall')

print('\n')

print('Overall Analysis of Germany in each year')
Fetch_Data_Medal('Overall', 'Germany')

print('\n')

print('Overall Analysis of all the countries in 1900')
Fetch_Data_Medal(1900, 'Overall')
```
